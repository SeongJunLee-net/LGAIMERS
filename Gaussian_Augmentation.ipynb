```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./preprocessing_train.csv')
train
```

```python
condition = (train['target'] == 'AbNormal')
condition
```

```python
float_columns = train.columns[((train.dtypes==float)|(train.dtypes==int)).values]
object_columns = train.columns[~((train.dtypes==float)|(train.dtypes==int)).values]
print(len(float_columns))
print(len(object_columns))
```

```python
import time
```

```python
seed = 42
np.random.seed(seed)
data_times = 5
augmentated_row_list = []
tmp_train = train.copy()
start = time.time()
for index,data in train.loc[condition,:].iterrows():
    cat_item = data[object_columns]
    for i in range(data_times):
        float_item = data[float_columns] + np.random.randn(len(float_columns))
        concat_item = pd.concat([cat_item,float_item]).to_frame().T
        tmp_train = pd.concat([tmp_train,concat_item],ignore_index=True, axis=0)
tmp_train
```

```python
tmp_train.to_csv('gausian_augmented_train.csv',index=False)
```
