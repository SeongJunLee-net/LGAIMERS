```python
condition = (train.dtypes=='object')
object_columns = list(train.columns[condition])
effective_columns = ['Chamber Temp. Judge Value_AutoClave','Equipment_Dam','Equipment_Fill2','Equipment_Fill1']
object_columns = [col for col in object_columns if col not in effective_columns]
object_columns.remove('target')
object_columns
```

```python
train.drop(columns = object_columns,inplace = True)
test.drop(columns = object_columns,inplace = True)
```

```python
encoder = OneHotEncoder(sparse_output=False)
one_hot_train = encoder.fit_transform(train[effective_columns])
one_hot_train = pd.DataFrame(one_hot_train,columns=encoder.get_feature_names_out())

poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
crossed_train = poly.fit_transform(one_hot_train)
train_cat_df = pd.DataFrame(crossed_train,columns=poly.get_feature_names_out())


one_hot_test = encoder.transform(test[effective_columns])
one_hot_test = pd.DataFrame(one_hot_test,columns=encoder.get_feature_names_out())
crossed_test = poly.fit_transform(one_hot_test)
test_cat_df = pd.DataFrame(crossed_test,columns=poly.get_feature_names_out())
```
